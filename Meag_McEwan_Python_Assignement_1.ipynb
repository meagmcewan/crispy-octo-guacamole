```python
import pandas as pd
import numpy as np
```

Read CSV file

```python
df = pd.read_csv('Travel_Pony_Facebook.csv')
```

```python
df.head()
```

Creating Cost Per Impression column

```python
df['Cost Per Impression'] = df['Amount Spent (USD)'] / df['Impressions']
```

Importing datetime and calendar to change dates into days of the week
Then converting the Start Date Column so that datetime can read the dates and convert to days of the week

```python
import datetime
import calendar
```

```python
df['Start Date']= pd.to_datetime(df['Start Date'])
df['Day of the Week'] = df['Start Date'].dt.day_name()
```

Getting general stats for the database

```python
df.describe()
```

Take average cost per impression grouped by Day of the Week.
Friday has the highest cost, Saturday has the lowest cost (although this still includes ads that have $0 for amount spent, so that is pulling down average costs)

```python
df_DAY = df.groupby('Day of the Week')['Cost Per Impression'].mean()
df_DAY
```

Correlations between Amount Spent and: Reach, Frequency, Unique Clicks, and Page Likes

Amount Spent has the highest correlation with Unique Clicks meaning as amount spent increases, unique clicks will increase the most out of the other factors along with it.

Mulitple regression analysis - Unique Clicks as dependent variable, Frequency and Reach as independent variables

```python
import statsmodels.api as sm
```

```python
x = df[['Frequency', 'Reach']]
y = df['Unique Clicks']

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

model.summary()
```

Resgression analysis of variable serarately - Reach is the stronger predictor of Unique Clicks, meaning variations in Reach better predict variations in Unique Clicks.

```python
x = df['Reach']
y = df['Unique Clicks']

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

model.summary()
```

```python
x = df['Frequency']
y = df['Unique Clicks']

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

model.summary()
```
