Scenario - create a dictionary with respondents answers to knowledge about different programing languages

```python
# print all the outputs in a cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
```

```python
respondents={'Amy': 0,
 'Bob': 1,
 'Catherine': 2,
 'Dorothy': 3,
 'Edward': 4,
 'Frankie': 5,
 'Gregory': 6,
 'Heather': 7,
 'Iris': 8,
 'Javard': 9}
```

```python
survey=['I have basic knowledge in Python.',
 'I am proficient in Java, and has some knowledge in R.',
 'I am familiar with C++, familiar with R, and familiar with Matlab.',
 'I have never used Python, but proficient in R.',
 'I have some knowledge in C++, and familiar with SQL.',
 'I have never used C++, and have basic knowledge in Python.',
 'I have never used Java, and have basic knowledge in SQL.',
 'I am proficient in C++, and proficient in Python.',
 'I have basic knowledge in R, and some knowledge in C++, but never used Python.',
 'I am familiar with Tableau, and familiar with Excel but never used R.']
```

put dictionary of respondents and survey answers into a dataframe

```python
df = pd.DataFrame({'Respondents':respondents})
df['survey'] = survey
df
```

assign numerical values to level of prficiency for each language in survey responses

```python
df['R'] = df['survey'].apply(lambda x : 4 if 'proficient in R' in x else 3 if 'familiar with R' in x else 2 if 'basic knowledge in R' in x else 1 if 'some knowledge in R' in x else 0)
df['CPP'] = df['survey'].apply(lambda x : 4 if 'proficient in C++' in x else 3 if 'familiar with C++' in x else 2 if 'basic knowledge in C++' in x else 1 if 'some knowledge in C++' in x else 0)
df['Python'] = df['survey'].apply(lambda x : 4 if 'proficient in Python' in x else 3 if 'familiar with Python' in x else 2 if 'basic knowledge in Python' in x else 1 if 'some knowledge in Python' in x else 0)
df['SQL'] = df['survey'].apply(lambda x : 4 if 'proficient in SQL' in x else 3 if 'familiar with SQL' in x else 2 if 'basic knowledge in SQL' in x else 1 if 'some knowledge in SQL' in x else 0)
df['Matlab'] = df['survey'].apply(lambda x : 4 if 'proficient in Matlab' in x else 3 if 'familiar with Matlab' in x else 2 if 'basic knowledge in Matlab' in x else 1 if 'some knowledge in Matlab' in x else 0)
df['Tableau'] = df['survey'].apply(lambda x : 4 if 'proficient in Tableau' in x else 3 if 'familiar with Tableau' in x else 2 if 'basic knowledge in Tableau' in x else 1 if 'some knowledge in Tableau' in x else 0)
df['Java'] = df['survey'].apply(lambda x : 4 if 'proficient in Java' in x else 3 if 'familiar with Java' in x else 2 if 'basic knowledge in Java' in x else 1 if 'some knowledge in Java' in x else 0)
df['Excel'] = df['survey'].apply(lambda x : 4 if 'proficient in Excel' in x else 3 if 'familiar with Excel' in x else 2 if 'basic knowledge in Excel' in x else 1 if 'some knowledge in Excel' in x else 0)
df
```

drop the text survey responses

```python
df.drop(['survey'], axis = 1, inplace=True)
```
